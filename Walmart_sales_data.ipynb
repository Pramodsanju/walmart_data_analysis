```python
import matplotlib.pyplot as plt

# Split dataset into holiday and non-holiday weeks
holiday_sales = data[data["Holiday_Flag"] == 1]["Weekly_Sales"].mean()
non_holiday_sales = data[data["Holiday_Flag"] == 0]["Weekly_Sales"].mean()

# Create a comparison DataFrame
comparison_df = pd.DataFrame({
    "Category": ["Holiday", "Non-Holiday"],
    "Average Weekly Sales": [holiday_sales, non_holiday_sales]
})

# Plot comparison
plt.figure(figsize=(8, 5))
plt.bar(comparison_df["Category"], comparison_df["Average Weekly Sales"], color=["red", "blue"], edgecolor="black")
plt.title("Comparison of Weekly Sales: Holiday vs. Non-Holiday Weeks")
plt.ylabel("Average Weekly Sales ($)")
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Show plot
plt.show()

# Display comparison DataFrame
comparison_df
```

```python
# Filter for the first 20 stores
top_20_stores = data["Store"].unique()[:20]

# Count the number of records per store (first 20 stores only)
store_counts_20 = data[data["Store"].isin(top_20_stores)]["Store"].value_counts().sort_index()

# Plot the number of entries per store (first 20 stores)
plt.figure(figsize=(8, 5))
store_counts_20.plot(kind="bar", color="skyblue")
plt.title("Number of Weekly Records per Store (First 20 Stores)")
plt.xlabel("Store ID")
plt.ylabel("Count of Records")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Compute total sales per store (first 20 stores only)
store_sales_20 = data[data["Store"].isin(top_20_stores)].groupby("Store")["Weekly_Sales"].sum().sort_values(ascending=False)

# Plot total sales per store (first 20 stores)
plt.figure(figsize=(10, 6))
store_sales_20.plot(kind="bar", color="red")
plt.title("Total Sales per Store (First 20 Stores)")
plt.xlabel("Store ID")
plt.ylabel("Total Sales ($)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Count the number of holiday vs. non-holiday weeks
holiday_counts = data["Holiday_Flag"].value_counts()

# Labels for the pie chart
labels = ["Non-Holiday Weeks", "Holiday Weeks"]

# Plot the pie chart
plt.figure(figsize=(8, 6))
plt.pie(
    holiday_counts,
    labels=labels,
    autopct="%1.1f%%",
    startangle=140,
    colors=["blue", "red"]
)

# Add a title
plt.title(f"Distribution of Holiday vs. Non-Holiday Weeks (Total Weeks: {data.shape[0]})")

# Show plot
plt.show()
```
